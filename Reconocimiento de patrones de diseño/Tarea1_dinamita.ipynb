(b) f (x) = cos(x)

```python
# Definición de la función f(x) y su derivada df(x)
def f(x):
    return math.cos(x)

def df(x):
    return -math.sin(x)

# Implementación del método de Newton-Raphson
def newton_raphson(f, df, x0, tol, max_iter):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        if abs(fx) < tol:
            return x
        dfx = df(x)
        if dfx == 0:
            break
        x = x - fx/dfx
    return None

# Especificación de parámetros
x0 = 2  # Valor inicial arbitrario
tol = 1e-8
max_iter = 1000

# Aplicación del método de Newton-Raphson
x = newton_raphson(f, df, x0, tol, max_iter)

# Impresión de la aproximación de la raíz encontrada
if x is not None:
    print("Aproximación de la raíz de cos(x) = 0:")
    print(x)
else:
    print("No se pudo encontrar una solución.")
```

4. Usando python y la ecuación de la recta, grafica la recta que pasa por los puntos, además indica la pendiente de la recta:

```python
def graficarPuntos(x1, y1, x2, y2):
    """
    Grafica la recta que pasa por dos puntos dados.

    Args:
        x1 (float): La primera entrada del primer punto.
        y1 (float): La segunda entrada del primer punto.
        x2 (float): La primera entrada del segundo punto.
        y2 (float): La segunda entrada del segundo punto.

    Returns:
        None

    Muestra un gráfico con la recta que pasa por los puntos dados.

    """
    # Definir los puntos
    punto1 = np.array([x1, y1])
    punto2 = np.array([x2, y2])

    # Calcular la pendiente de la línea
    m = (punto2[1] - punto1[1]) / (punto2[0] - punto1[0])

    # Generar valores para los ejes X e Y
    ejex = np.linspace(min(punto1[0], punto2[0]) - 1, max(punto1[0], punto2[0]) + 1, max(punto1[0], punto2[0]) - min(punto1[0], punto2[0]) + 3)
    ejey = m * (ejex - punto1[0]) + punto1[1]

    # Crear la figura y los ejes
    fig, ax = plt.subplots()

    # Graficar los puntos
    ax.scatter(punto1[0], punto1[1], color='#f8766d', marker='o')
    ax.scatter(punto2[0], punto2[1], color='#f8766d', marker='o')

    # Graficar la línea
    plt.plot(ejex, ejey, color='#00bfc4')

    # Agregar ejes X e Y
    ax.axhline(0, color='black', linewidth=0.5)  # Eje horizontal (Y)
    ax.axvline(0, color='black', linewidth=0.5)  # Eje vertical (X)

    # Agregar título
    plt.title(f'Recta que pasa por los puntos: {(punto1[0], punto1[1])}, {(punto2[0], punto2[1])} con pendiente: {m}')

    # Mostrar el gráfico
    plt.show()
```

5. Genere una Interpolación Lineal usando $n$ puntos de la función $\cos(x)$ y $\tan(x)$. Sobre el intervalo $[0, 2π]$. Indica cuántos
puntos $n$ usaste. (Procura optimizar tu resultado).

```python
def interpolacionLineal(funcion, nombre, n, a = 0, b = 2*np.pi):
    """
    Realiza una interpolación lineal de una función en un intervalo dado.

    Args:
        funcion (función): La función cuya interpolación deseas realizar.
        nombre (str): El nombre de la función.
        n (int): El número de puntos conocidos para la interpolación.
        a (float): El punto inicial del intervalo.
        b (float): El punto final del intervalo.

    Returns:
        None

    Muestra un gráfico que representa la interpolación lineal de la función
    en el intervalo [a, b], utilizando los valores conocidos de la función.

    """
    # n puntos conocidos
    x = np.linspace(a, b, n)
    y = funcion(x)

    # Eje x para la interpolacion
    x_new = np.linspace(a, b, 100)

    # Interpolación lineal
    y_new = np.interp(x_new, x, y)

    # Grafica
    plt.plot(x_new,y_new, label='Valores Interpolados')
    plt.plot(x,y, 'o', label='Valores conocidos')
    plt.plot(x_new, funcion(x_new), '--', label=nombre)
    plt.title(f"Interpolación Lineal")
    plt.legend()
    plt.show()
```

```python
interpolacionLineal(lambda x : np.cos(x), 'Coseno', n = 10)
```

6. Genere una Interpolación Polinómica de Newton usando n puntos de la función cos(x) y tan(x). Sobre el intervalo [0, 2π].
Indica cuantos puntos n usaste.(Procura optimizar tu resultado).

```python
def interpolacion_newton(f,titulo, n, a=0, b=2*np.pi):
  x = np.array(np.linspace(a, b, n))
  y = f(x)

  def divididas_diferencias(x, y):
      n = len(x)
      coef = y.copy()
      for j in range(1, n):
          for i in range(n-1, j-1, -1):
              coef[i] = (coef[i] - coef[i-1]) / (x[i] - x[i-j])
      return coef

  coef = divididas_diferencias(x, y)

  def newton_interpolation(x, y, coef, x_new):
      n = len(x)
      p = coef[n-1]
      for i in range(n-2, -1, -1):
          p = coef[i] + (x_new - x[i]) * p
      return p

  x_new = np.linspace(a, b, 100)
  y_new = [newton_interpolation(x, y, coef, xi) for xi in x_new]

  plt.plot(x_new, y_new, label='Polinomio de Newton')
  plt.legend()

  plt.plot(x, y, 'o', label='Valores conocidos(Interpolación Lineal)')
  plt.legend()
  plt.xlabel('x')
  plt.ylabel('$p_i(x)$')
  plt.title('Polinomio de Newton')
  plt.show()
```

```python
interpolacion_newton(lambda x : np.cos(x), 'Coseno', 10)
```

```python
interpolacion_newton(lambda x : np.tan(x), 'Tangente', 10)
```

7. Genere una Interpolación Polinómica de Lagrange usando n puntos de la función cos(x) y tan(x). Sobre el intervalo [0, 2π].
Indica cuantos puntos n usaste y el polinomio obtenido. (Procura optimizar tu resultado).

```python
import sympy as sym

def interpolacion_lagrange(f,titulo, n, a=0, b=2*np.pi):
  xi = np.array(np.linspace(a, b, n))
  fi = f(xi)

  # PROCEDIMIENTO
  # Polinomio de Lagrange
  x = sym.Symbol('x')
  polinomio = 0
  divisorL = np.zeros(n, dtype = float)
  for i in range(0,n,1):
      # Termino de Lagrange
      numerador = 1
      denominador = 1
      for j  in range(0,n,1):
          if (j!=i):
              numerador = numerador*(x-xi[j])
              denominador = denominador*(xi[i]-xi[j])
      terminoLi = numerador/denominador

      polinomio = polinomio + terminoLi*fi[i]
      divisorL[i] = denominador

  # simplifica el polinomio
  polisimple = polinomio.expand()
  # para evaluación numérica
  px = sym.lambdify(x,polisimple)
  # Puntos para la gráfica
  muestras = 100
  pxi = np.linspace(a,b,muestras)
  pfi = px(pxi)

  # SALIDA
  print('Polinomio de Lagrange: ')
  print(polisimple)

  # Gráfica
  plt.plot(xi,fi,'o-', label = 'Puntos')
  plt.plot(pxi,pfi, label = 'Polinomio')
  plt.legend()
  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.title('Interpolación Lagrange')
  plt.show()
```

```python
interpolacion_lagrange(lambda x : np.cos(x), 'Coseno', 10)
```

```python
interpolacion_lagrange(lambda x : np.tan(x), 'Tangente', 10)
```
